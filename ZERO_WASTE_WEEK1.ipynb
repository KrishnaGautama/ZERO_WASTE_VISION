```python
import numpy as np

#RANDOM MATRIX
random_matrix = np.random.randint(0, 256, size=(15, 15))
print("Original Random Matrix:")
print(random_matrix)

# PROBLEM 1:TASK 1
threshold = 120
thresholded_matrix = np.where(random_matrix < threshold, 0, random_matrix)
print("\nThresholded Matrix (values < 120 set to 0):")
print(thresholded_matrix)

#PROBLEM 1: TASK 2
def print_boundary_clockwise(matrix):
    rows, cols = matrix.shape
    boundary_elements = []
    
    # Top row
    boundary_elements.extend(matrix[0, :])
    
    # Right column 
    boundary_elements.extend(matrix[1:rows-1, cols-1])
    
    # Bottom row in reverse order
    boundary_elements.extend(matrix[rows-1, ::-1])
    
    # Left column in reverse order 
    boundary_elements.extend(matrix[1:rows-1, 0][::-1])
    
    print("\nBoundary elements in clockwise order:")
    print(boundary_elements)

print_boundary_clockwise(random_matrix)

#PROBLEM 1: TASK 3
even_odd_binary = np.where(random_matrix % 2 == 0, 1, 0)
print("\nEven-Odd Binary Matrix (even = 1, odd = 0):")
print(even_odd_binary)
```

# PROBLEM 2

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

#TASK 1
arr = np.random.randint(0, 256, size=(20, 20), dtype=np.uint8)
print("Original Array (20x20):\n", arr)

# Row-wise stats
print("\nRow-wise Max:\n", np.max(arr, axis=1))
print("Row-wise Min:\n", np.min(arr, axis=1))
print("Row-wise Median:\n", np.median(arr, axis=1))

# Column-wise stats
print("\nColumn-wise Max:\n", np.max(arr, axis=0))
print("Column-wise Min:\n", np.min(arr, axis=0))
print("Column-wise Median:\n", np.median(arr, axis=0))

#TASK 2
resized = cv2.resize(arr, (200, 200), interpolation=cv2.INTER_NEAREST)

# Show using matplotlib
plt.figure(figsize=(4, 4))
plt.title("Resized Image (200x200) - Nearest Neighbor")
plt.imshow(resized, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()

#TASK 3
# Horizontal Gradient: left to right
horizontal_gradient = np.tile(np.linspace(0, 255, 256, dtype=np.uint8), (256, 1))

# Vertical Gradient: top to bottom
vertical_gradient = np.tile(np.linspace(0, 255, 256, dtype=np.uint8).reshape(256, 1), (1, 256))

# Display gradients
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title("Horizontal Gradient")
plt.imshow(horizontal_gradient, cmap='gray', vmin=0, vmax=255)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Vertical Gradient")
plt.imshow(vertical_gradient, cmap='gray', vmin=0, vmax=255)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

#USING MATLAB TO SHOW IMAGE IN JUPYTER NOTEBOOK
def show_image(image, title="", cmap_type='gray'):
    plt.figure(figsize=(5, 5))
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()

#TASK 1

image_path = "GREYSCALE.jpg"  
gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
inverted = 255 - gray  
show_image(gray, "Original Grayscale")
show_image(inverted, "Inverted Grayscale")
```

```python
#TASK 2
#USING MATLAB TO SHOW IMAGE IN JUPYTER NOTEBOOK
def show_image(image, title=""):
    plt.figure(figsize=(5, 5))
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_image)
    plt.title(title)
    plt.axis('off')
    plt.show()


image_path = "Color_Image.jpg"  # Update if needed

color_img = cv2.imread(image_path)


B, G, R = cv2.split(color_img)

# Zero the Red channel
R[:] = 0

# Merge back the modified image
no_red_img = cv2.merge([B, G, R])

show_image(color_img, "Original Color Image")
show_image(no_red_img, "Image with Red Channel Zeroed")
```

```python
def show_image(image, title=""):
    plt.figure(figsize=(5, 5))
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_image)
    plt.title(title)
    plt.axis('off')
    plt.show()
image_path = "Color_Image.jpg"
color_img = cv2.imread(image_path)


print("Image shape (Height, Width, Channels):", color_img.shape)

# Split into channels
B, G, R = cv2.split(color_img)

print("\n Blue channel sample:\n", B[:5, :5])
print("\n Green channel sample:\n", G[:5, :5])
print("\n Red channel sample:\n", R[:5, :5])

show_image(color_img, "Original Color Image")

#TASK 2
gray_manual = 0.299 * R + 0.587 * G + 0.114 * B
gray_manual = gray_manual.astype(np.uint8)

plt.figure(figsize=(5, 5))
plt.imshow(gray_manual, cmap='gray')
plt.title("Manual Grayscale Image")
plt.axis('off')
plt.show()

#TASK 3
white_pixels = np.sum(np.all(color_img == [255, 255, 255], axis=-1))
print(f"❄️ Number of pure white pixels: {white_pixels}")
```

# PROBLEM 5

```python
def show_image(image, title="", cmap_type=None):
    plt.figure(figsize=(5, 5))
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()

image_path = "Color_Image.jpg"
color_img = cv2.imread(image_path)


#  Task 1: Grayscale + Gaussian Blur

gray = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (9, 9), 0)

show_image(gray, "Grayscale Image", cmap_type='gray')
show_image(blurred, "Blurred Grayscale Image", cmap_type='gray')


#  Task 2: Subtract blurred image from white background
# 
white_bg = np.full_like(blurred, 255) 
inverted = cv2.subtract(white_bg, blurred)

show_image(inverted, "Inverted (White - Blurred)", cmap_type='gray')

# Task 3: Draw shapes on a blank (512x512) canvas

canvas = np.zeros((512, 512, 3), dtype=np.uint8)

#  Blue rectangle
cv2.rectangle(canvas, (50, 50), (462, 462), color=(255, 0, 0), thickness=3)

#  Green filled circle
cv2.circle(canvas, (300, 300), radius=50, color=(0, 255, 0), thickness=-1)

#  Red diagonal line (top-right to bottom-left)
cv2.line(canvas, (512, 0), (0, 512), color=(0, 0, 255), thickness=2)

# Show final canvas
show_image(canvas, "Canvas with Shapes")
```
